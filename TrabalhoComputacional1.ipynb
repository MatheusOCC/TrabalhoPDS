## 1. Calcule e plote a transformada de Fourier do sinal $x[n] = x_1[n] + x_2[n] + x_3[n]$ :

$$x_1[n] = cos(0,2pin)$$
$$x_2[n] = cos(0,4pin - pi/2)$$
$$x_3[n] = cos(0,8pin + pi/5)$$

- O eixo da magnitude do espectro de magnitude deve expressar $|(X(e^jw)|$;
- O espectro deve ser plotado em funçãoda frequência normalizada;
- O espectro de fase deve indicar o valor principal;
- Repita a plotagem acima, somente para o espectro de magnitude,considerando uma frequência de amostragem fs = 20kHz e gerando o eixo de frequências em Hz equivalente.

**Explicando o primeiro passo**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Primeiro passo: 

fs = 40 # frequência de amostragem
ps = 1/fs # período de amostragem
n = np.arange(0,40,ps) # 1600 amostras
t = n*ps # tempo

x1 = np.cos(0.2*np.pi*n) # x1 = np.cos(0.1*2*np.pi*n) 
x2 = np.cos(0.4*np.pi*n - np.pi/2) # x2 = np.cos(0.2*2*np.pi*n + np.pi/2)
x3 = np.cos(0.8*np.pi*n + np.pi/5) # x3 = np.cos(0.4*2*np.pi*n + np.pi/5)

x = x1 + x2 + x3

```

**Explicando o segundo passo**

```python
plt.figure(1)
plt.title("Sinal Original I")
plt.ylabel("Amplitude")
plt.xlabel("Tempo (s)")
plt.plot(t,x)
```

**Explicando o terceiro passo**

```python
fft_transformada = np.fft.fft(x)
freq = np.fft.fftfreq(len(fft_transformada),ps)

plt.figure(2)
plt.title("Sinal da Transfomada de Fourier de x[n] I")
plt.ylabel("Magnitude")
plt.xlabel("Frequência (Hz)")
plt.plot(freq[freq > 0],np.abs(fft_transformada)[freq > 0])
```

**Explicando o quarto passo**

```python
N = len(n)
fn = np.arange(0,N)

plt.figure(3)
plt.title("Normalizando o Sinal da Transfomada de Fourier de x[n] I")
plt.ylabel("Magnitude")
plt.xlabel("Frequência normalizada")
plt.plot(fn[freq > 0]*2/Nnp.abs(fft_transformada)[freq > 0])
```

**Explicando o quinto passo**

```python
plt.figure(4)
plt.title("Fase do Sinal da Transfomada de Fourier de x[n] I")
plt.plot(freq,np.angle(fft_transformada))
```

**Concluindo**

## 2. Calcule e plote a transformada de Fourier do sinal $x[n] = x_1[n −M −1] + x_2[n −2M −2] + x_3[n]$:

$$x1[n] = w[n]cos(0,2πn)$$
$$x2[n] = w[n]cos(0,4πn −π/2)$$
$$x3[n] = w[n]cos(0,8πn + π/5)$$

Considere que w[n] é uma janela de Hamming de comprimento $(M + 1) = 61$ definida por $w[n] = 0,54 −0,46cos(2πn/M), 0 ≤n ≤M$ e zero fora desse intervalo. Para realizar a plotagem, siga as seguintes especificações:

- O eixo da magnitude do espectro de magnitude deve expressar $|(X(e^jw)|$;
- O espectro deve ser plotado em funçãoda frequência normalizada;
- O espectro de fase deve indicar o valor principal;
- Repita a plotagem acima, somente para o espectro de magnitude,considerando uma frequência de amostragem fs = 20kHz e gerando o eixo de frequências em Hz equivalente.

**Explicando o primeiro passo**

```python
fs = 40 # frequência de amostragem
ps = 1/fs # período de amostragem
n = np.arange(0,40,ps)
n1 = np.arange(0,1,ps) # 40 amostras
n1_1 = np.pad(n1, (0,len(n) - len(n1)),'constant')
t = n*ps # tempo
M = 60

wn =  0.54 - 0.46 * np.cos((2*np.pi*n1_1)/M)

x1 = wn * np.cos(0.2*np.pi*(n-M-1))
x2 = wn * np.cos(0.4*np.pi*(n - 2*M - 2) - np.pi/2)
x3 = wn * np.cos(0.8*np.pi*n + np.pi/5)

x = x1 + x2 + x3
```

**Explicando o segundo passo**

```python
plt.figure(1)
plt.title("Sinal Original I")
plt.ylabel("Amplitude")
plt.xlabel("Tempo (s)")
plt.plot(t,x)
```

**Explicando o terceiro passo**

```python
fft_transformada = np.fft.fft(x)
freq = np.fft.fftfreq(len(fft_transformada),ps)

plt.figure(2)
plt.title("Sinal da Transfomada de Fourier de x[n] I")
plt.ylabel("Magnitude")
plt.xlabel("Frequência (Hz)")
plt.plot(freq[freq > 0],np.abs(fft_transformada)[freq > 0])
```

**Explicando o quarto passo**

```python
N = len(n)
fn = np.arange(0,N)
plt.figure(3)
plt.title("Normalizando o Sinal da Transfomada de Fourier de x[n] I")
plt.ylabel("Magnitude")
plt.xlabel("Frequência normalizada")
plt.plot(fn[freq > 0]*2/N,np.abs(fft_transformada)[freq > 0])
```

**Explicando o quinto passo**

```python
plt.figure(4)
plt.title("Fase do Sinal da Transfomada de Fourier de x[n] I")
plt.plot(freq,np.angle(fft_transformada))
```
